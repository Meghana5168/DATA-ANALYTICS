#### Series data structure

```python
# Series is a 1D array.
# In Ml we store the "Target" in the dataset as a series
```

```python
# empty series 
ser=pd.Series()
print(ser)
```

```python
# Series from the pyhton List
list1=[1,2,3,4,5,6]
ser1=pd.Series(list1)
print(ser1)
```

```python
#create series from the numpy array
import numpy as np
data= np.array(["Meghana","Pikki","nikki"])
type(data)
ser2=pd.Series(data)
print(ser2)
type(ser2)
```

```python
print(ser2[0])
```

```python
#print(ser2[-3])
# pandas series dont support the negative indexing
```

```python
data= np.array(["Meghana","Pikki","nikki","chikki"])
ser2=pd.Series(data, index=['a','b','c','d'])
ser2
```

```python
print(ser2['a'])
```

```python
# convert the dictionary into pandas series
data1={"Meghana":21,"Pikki":20,"nikki":35,"chikki":45}
ser3= pd.Series(data1)
ser3
```

```python
# convert the tuple into pandas series
data3=(5,6,7,8,9)
ser4=pd.Series(data3)
ser4
```

```python
# convert the set into pandas series
data4={"Meghana","Pikki","nikki","chikki"}
ser5=pd.Series(list(data4))
ser5
# Here set is an unordered datatype, so it won't convert directly into series
```

```python
# convert the scaler to pandas series
val=100
ser6=pd.Series(val)
ser6
ser6=pd.Series(val, index = ['a','b','c'])
ser6

# we convert the scaler(means single value) to the pandas series
```

# Data Frame

```python
# DataFrame is a 2D data structure
# we will use DataFrame to Store Datasets
```

```python
#empty dataframe
df=pd.DataFrame()
print(df)
```

```python
# convert the list to dataframe
data=["apple","banana","mango"]
df1=pd.DataFrame(data)
df1
```

```python
# convert the 2D list to dataframe
data1=[["apple",250],["banana",80],["mango",300]]
df=pd.DataFrame(data1,columns=["Fruit Name","price"])
print(df)
```

```python
# convert the 2D tuple to dataframe
data1=(("apple",250),("banana",80),("mango",300))
df=pd.DataFrame(data1,columns=["Fruit Name","price"])
print(df)
```

```python
# convert the 2D set to dataframe
data1={{"apple",250},{"banana",80},{"mango",300}} 
df=pd.DataFrame(data1,columns=["Fruit Name","price"])
print(df)
```

```python
# create a data frame from the dictionary
data1={ 
    "name":["pappu","briyani","curd","desert"],
    "count":[1,5,100,10]
}
df=pd.DataFrame(data1)
df
```

```python
data1={"Meghana":21,"Pikki":20,"nikki":35,"chikki":45}
data2={
    "Name":["Meghana","Pikki","nikki","chikki"],
    "Age":[21,20,35,45]
}
df2=pd.DataFrame(data2)
df2
```

```python
#convert the series into dataframe
sre=pd.Series([1,2,3,4,5],name="count")
df=sre.to_frame()
df
type(df)
```
